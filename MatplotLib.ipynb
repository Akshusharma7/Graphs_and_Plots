```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(1,5,11)
y = x**2
x
```

```python
y
```

```python
#FUNCTIONAL WAY OF PLOTING
plt.plot(x,y, 'r-')
plt.xlabel('X Lable')
plt.ylabel('Y Lable')
plt.title('tilte')
```

```python
#Object oriented Method

fig = plt.figure()
axces1 = fig.add_axes([0.1, 0.1, 0.9, 0.9])
axces2 = fig.add_axes([0.2, 0.6, 0.4, 0.3])

axces1.plot(x,y)
axces1.set_title("Larger Plot")
axces2.plot(y,x)
axces2.set_title("Smaller Plot")
```

```python
fig,axes = plt.subplots(nrows = 1, ncols= 2)
plt.tight_layout()
```

```python
axes# it is just a list of matplotlib we can itterate over it
```

```python
for current_ax in axes:
    current_ax.plot(x,y)

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.show()
```

### Figure size and DPI

```python
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize= (8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
#Saves the file in current directory
fig.savefig("my_fig.png", dpi=200)
```

```python
#Legends

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label = 'X Square')
ax.plot(x,x**3, label = 'X Cube')

ax.legend(loc = 0)
```

### Plot Apperance

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color="red" ,linewidth=3, alpha=0.5)# You can add RGB hex color code. Take it from Google(#f1f442 - yellow)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y**2 , color = 'green', lw = 3,linestyle ="-", marker="o", markersize=10) #linestyle= "--" or "-." or "+-" or "++" "steps"etc
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y**2 , color = 'green', lw = 3,linestyle ="-", marker="o", markersize=10, markerfacecolor="red",
       markeredgewidth = 3, markeredgecolor= "yellow")
```
